**Outliers Treatment**

*a) Checking how many values are > or < (mean +-3*std)

```python
for col in iris_data.columns.tolist()[:-1]:
    print iris_data[col][(iris_data[col] > (iris_data[col].mean() + 3*iris_data[col].std())) | (iris_data[col] < (iris_data[col].mean() - 3*iris_data[col].std()))]
    
```

*b)Checking how many values are > 95% percentile or < 5% percentile*

```python
for col in iris_data.columns.tolist()[:-1]:
    iris_data[col][(iris_data[col] > np.percentile(iris_data[col],95)) | (iris_data[col] < np.percentile(iris_data[col],5))]
    
```

*We will go with approach 'a' to identify outliers as it gives lesser outliers than percentile approach. Outliers are usually replaced with column median*

```python
for col in iris_data.columns.tolist()[:-1]:
    iris_data[col].replace( iris_data[col][iris_data[col]>iris_data[col].mean()+3*iris_data[col].std()],  iris_data[col].median(), inplace = True)
    iris_data[col].replace( iris_data[col][iris_data[col]<iris_data[col].mean()-3*iris_data[col].std()],  iris_data[col].median(), inplace = True)
    
    
```

*There are no missing values in the dataset, so no need for Missing value treatment*

#### Data Modelling

```python
X = iris_data.loc[:, iris_data.columns != 'class']
y = iris_data['class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
def modelfit(alg, X_train, y_train, y_test, featureImp = True):
    alg.fit(X_train, y_train)
    predictions = alg.predict(X_test)
    print metrics.accuracy_score(y_test, predictions)
    
    if featureImp:
        feat_imp = pd.Series(alg.feature_importances_, X_train.columns.tolist()).sort_values(ascending = True)
        feat_imp.plot(kind = 'bar', title = 'Feature Importance')
    
```

```python
alg = LogisticRegression()
modelfit(alg, X_train, y_train, y_test, False)
```
