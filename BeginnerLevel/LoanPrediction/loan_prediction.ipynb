```python
test_data = pd.read_csv("test_data.csv", index_col = 'Loan_ID')
```

```python
test_data.dtypes
```

```python
test_data.apply(lambda x: sum(x.isnull()), axis = 0)
```

```python
test_data['TotalIncome'] = test_data['ApplicantIncome'] + test_data['CoapplicantIncome']
test_data.drop(['ApplicantIncome','CoapplicantIncome'], axis = 1, inplace = True)
```

```python
def replace_outliers_with_median(df,col):
    return df[col].replace( df[df[col]>(df[col].mean()+3*df[col].std())][col] , df[col].median(), inplace = True)
```

```python
for x in ['Female','Male']:
    df = test_data[test_data['Gender']==x]
    for col in ['LoanAmount','TotalIncome']:
        replace_outliers_with_median(df, col)
    test_data[test_data['Gender']==x] = df
```

```python
def replace_missing_values_numericvariable(df,col):
    return df[col].fillna(df[col].median(), inplace = True)
    
def replace_missing_values_categoricalvariable(df,col):
    return df[col].fillna(df[col].value_counts().index[0], inplace = True)
```

```python
for x in ['Female','Male']:
    df = test_data[test_data['Gender']==x]
    for col in test_data.select_dtypes(include = ['float64']).columns.tolist():
        replace_missing_values_numericvariable(df, col)
    test_data[test_data['Gender']==x] = df
    
for col in test_data.select_dtypes(include = ['object']).columns.tolist():
    replace_missing_values_categoricalvariable(test_data, col)
```

```python
def count_missing_values(x):
    return sum(x.isnull())

test_data.apply(count_missing_values, axis = 0)
```
